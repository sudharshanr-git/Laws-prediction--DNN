```python
#creating a framework which is used to encode

framework=dict()
for i in selected_laws:
    framework[i]=0
framework
```

```python
def change(laws):#takes the list of laws and outputs the encoded version
    l=framework.copy()
    for i in laws:
        if i in l.keys():
            l[i]=1
    return list(l.values())
```

```python
from sentence_transformers import SentenceTransformer

print("Loading the Sentence Transformer model...")
model = SentenceTransformer('LaBSE')
print("Model loaded successfully.")
```

```python
def para_to_sentences(in_para):
    para = in_para.split("\n")
    para = [i for i in para if len(i)>0]
    selected = []
    for i in para:

        if i[-1]==":" or len(i)==0:          #removing headings and if last line has \n
            continue

        words = i.split()
        if  len(words)<=1:          #removes number headings
            continue
        if any(True for i in ".)-" if i in words[1]):#removes numbering in numbers or alphabets as in first word
            selected.append(" ".join(words[2:]))
        elif any(True for i in ".)-" if i in words[0]):#removes numbering in numbers or alphabets as in second word
            selected.append(" ".join(words[1:]))
        else:
            selected.append(i)
    selected = [i for i in selected if len(i)>0]  #removes empty lines
    i=0
    while True:                 #deals with word hyphenization

        if selected[i][-1] == '-':
            selected[i] = selected[i][:-1]+selected[i+1]
            selected.pop(i+1)
            continue
        i+=1
        if len(selected)>=i:
            break

    selected = [i.replace("<nl>", " ")+'.' for i in '<nl>'.join(selected).split('.<nl>')]         #removes unwanted \n
    selected = list(dict.fromkeys(selected))                        #removes duplicates without changing the order
    return selected
```

```python
#for dnn - mean pooled

from tqdm import tqdm
fact_vectors=[]
index = 0
facts_error_docs = []
for i in tqdm(df['facts']):
    try:
        sentences = para_to_sentences(i)
    except Exception as e:
        fact_vectors.append([0]*768)
        facts_error_docs.append(index)
        index+=1
        continue
    fact_vectors.append(np.mean(model.encode(sentences), axis = 0).tolist())
    index+=1
np.save("fact_embeddings.npy",np.array(fact_vectors))
df['facts_encoded']=fact_vectors
```

```python
#for dnn - mean pooled

from tqdm import tqdm
considerations_vectors=[]
index = 0
considerations_error_docs = []
for i in tqdm(df['considerations']):
    try:
        sentences = para_to_sentences(i)
    except Exception as e:
        considerations_vectors.append([0]*768)
        considerations_error_docs.append(index)
        index+=1
        continue
    considerations_vectors.append(np.mean(model.encode(sentences), axis = 0).tolist())
    index+=1
np.save("considerations_embeddings.npy",np.array(considerations_vectors))
df['considerations_encoded']=considerations_vectors
```

```python
#for lstm

from tqdm import tqdm
fact_vectors=[]
index = 0
facts_error_docs = []
for i in tqdm(df['facts']):
    try:
        sentences = para_to_sentences(i)
    except Exception as e:
        fact_vectors.append([[0]*768])
        facts_error_docs.append(index)
        index+=1
        continue
    fact_vectors.append(model.encode(sentences).tolist())
    index+=1
np.save("fact_embeddings.npy",np.array(fact_vectors))
df['facts_encoded']=fact_vectors
```

```python
#for lstm

from tqdm import tqdm
considerations_vectors=[]
index = 0
considerations_error_docs = []
for i in tqdm(df['considerations']):
    try:
        sentences = para_to_sentences(i)
    except Exception as e:
        considerations_vectors.append([[0]*768])
        considerations_error_docs.append(index)
        index+=1
        continue
    considerations_vectors.append(model.encode(sentences).tolist())
    index+=1
np.save("considerations_embeddings.npy",np.array(considerations_vectors))
df['considerations_encoded']=considerations_vectors
```
